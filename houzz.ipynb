```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import csv
import os
links=[]
names=[]
websites=[]
descriptions=[]
addresses=[]
reviews=[]
phones=[]
review_dates=[]
total_reviews=[]
facebooks=[]
awards=[]
review_contents=[]

def altman():

    # Initialize WebDriver
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--enable-javascript")
    driver = webdriver.Chrome(options=chrome_options)
    driver.maximize_window()
    # driver.implicitly_wait(2)

    main_url = "https://www.houzz.com/professionals/query?l=california&tid=34257"

    driver.get(main_url)

    def scroll_to_bottom():
        try:
            # Scroll to the bottom of the page
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # print("Scrolled to the bottom successfully.")
        except Exception as e:
            print(f"Error while scrolling: {str(e)}")


    folder_path = "Output/Temporary"

    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_path}' created.")

    def create_csv():
        csv_file = "Output/Temporary/houzz.csv"
        with open(csv_file, mode="w", newline="", encoding="utf-8-sig") as file:
            fieldnames = ["Houzz Link", "Name","Company Website","Description","Address","Phone", "Review","Latest Review Date","Latest Review Content","Total Review","Facebook Link","Houzz Awards"]
            writer = csv.writer(file)
            writer.writerow(fieldnames)

            for row in zip(links,names,websites,descriptions,addresses,phones,reviews,review_dates,review_contents,total_reviews,facebooks,awards):
                writer.writerow(row)


    try:


        time.sleep(2)

        # while True:
        #     prev_height = driver.execute_script("return document.body.scrollHeight")
        #     scroll_to_bottom()
        #     time.sleep(5) 
        #     new_height = driver.execute_script("return document.body.scrollHeight")
        #     if new_height == prev_height:
        #         break
        
        elements = driver.find_elements(By.XPATH,"//a[@class='hz-pro-ctl']")

        print(len(elements))
        for element in elements:
            exp_list = []

            try:
                
                link = element.get_attribute('href')
                
                driver.execute_script("window.open(arguments[0], '_blank');", link)
                new_tab_handle = driver.window_handles[-1]
                driver.switch_to.window(new_tab_handle)
                try:
                    name = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//h1[@class='sc-mwxddt-0 dhSazp hz-track-me']"))
                    )
                    name= name.text
                    # name = driver.find_element(By.XPATH,"//h1[@class='sc-mwxddt-0 dhSazp hz-track-me']").text
                except: name = ""
                
                try: 
                    description = driver.find_element(By.XPATH,"//div[@class='sc-183mtny-0 fzynIk profile-about']").text
                    descriptions.append(description)
                except: descriptions.append("")

                try:
                    website = driver.find_element(By.XPATH,"//span[@class='sc-mwxddt-0 Website__EllipsisText-sc-19fzbgj-0 ctdszu dRuQKg']").text
                    # linkedins.append(linkedin_url.get_attribute('href'))
                except:
                    website.append("")  
                try:
                    address = driver.find_element(By.XPATH,"//h3[normalize-space()='Address']/./following-sibling::div").text
                except: address = ""
                addresses.append(address)
                try:
                    phone = driver.find_element(By.XPATH,"//h3[normalize-space()='Phone Number']/./following-sibling::div").text
                except:    phone=''
                phones.append(phone)  
                  
                try:
                    review = driver.find_element(By.XPATH,"//span[@class='ReviewAggregation__StyledRating-sc-11mmvxo-1 gghPyS']").text
                except:review=''
                reviews.append(review) 
                try:
                    review_date = driver.find_element(By.XPATH,"//div[@class='reviews-wrapper']//div[@class='review-item'][1]//div[@class='review-item__actions-panel review-item__time-panel text-xs']").text
                except:
                    review_date=''    
                try:
                    review_content = driver.find_element(By.XPATH,"//div[@class='reviews-wrapper']//div[@class='review-item'][1]//div[@class='review-item__body-string']").text
                except:
                    review_content=''      
                try:
                    reviews_list = driver.find_elements(By.XPATH,"//div[contains(@class,'reviews-wrapper')]/div[@class='review-item']")
                    total_review = len(reviews_list)
                except:
                    total_review='' 
                try:
                    facebook = driver.find_element(By.XPATH,"//a[@aria-label='Find me on Facebook']").get_attribute('href')
                except:
                    facebook='' 
                try:
                    award = driver.find_element(By.XPATH,"//h2[@id='credentials-label']/following-sibling::div//div[@class='side-bar-container'][1]/div[1]").text
                except:
                    award=''          


                driver.close()  # Close the current tab
                driver.switch_to.window(driver.window_handles[-1])     
            except Exception as e:
                print(e)
                link=""
                descriptions.append("")
                review_dates.append('')
                total_reviews.append('')
                facebooks.append('')
                awards.append('')
                websites.append('')
                review_contents.append('')
                

               

           
            

             
            review_dates.append(review_date)
            total_reviews.append(total_review)
            facebooks.append(facebook)
            awards.append(award)
            review_contents.append(review_content)
            links.append(link)
            names.append(name)
            websites.append(website)
                  


        
               
    except Exception as e:
        print(e)
        create_csv()

    finally:
        driver.quit()
        create_csv()


    create_csv()

altman()
```
